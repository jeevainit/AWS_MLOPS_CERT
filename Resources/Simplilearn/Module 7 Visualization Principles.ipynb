## Principles of Effective Data Visualization

### The Purpose of Data Visualization

Data visualization serves several crucial purposes:

1. **Exploration**: Discovering patterns, trends, and outliers in your data
2. **Analysis**: Examining relationships and testing hypotheses
3. **Communication**: Presenting findings to others in a clear, compelling way
4. **Decision-making**: Providing insights that inform actions

Effective visualizations help transform raw data into actionable insights by leveraging human visual perception. Our visual system can quickly detect patterns, anomalies, and relationships when data is presented graphically.

### Core Principles

These core principles will help you create effective visualizations:

#### 1. Know Your Purpose and Audience

Before creating a visualization, ask yourself:
- What question am I trying to answer?
- Who will be viewing this visualization?
- What actions should they take based on this information?

The needs of a data scientist exploring data differ from those of an executive making a strategic decision.

#### 2. Choose the Right Visualization Type

Different data relationships call for different visualization types:
- **Comparisons**: Bar charts, dot plots
- **Distributions**: Histograms, box plots
- **Compositions**: Pie charts, stacked bar charts
- **Relationships**: Scatter plots, bubble charts
- **Trends over time**: Line charts

We'll explore this in detail in the next section.

#### 3. Focus on Clarity and Simplicity

- Remove unnecessary elements (chart junk)
- Use clear titles and labels
- Don't make the viewer work to understand the message
- Simplify to the essential components

Edward Tufte, a pioneer in data visualization, introduced the concept of the "data-ink ratio" - maximize the ink used for displaying data and minimize non-data ink.

#### 4. Use Visual Hierarchy

Guide the viewer through your visualization:
- Emphasize what's important
- De-emphasize secondary elements
- Use size, color, and position strategically

#### 5. Be Honest and Accurate

- Don't distort the data
- Use appropriate scales
- Provide context
- Include uncertainty when relevant

#### 6. Use Color Effectively

- Use color with purpose, not decoration
- Be aware of color blindness issues
- Leverage color psychology
- Maintain consistency across related visualizations

#### 7. Make it Accessible

- Add proper labels and legends
- Use enough contrast
- Consider how the visualization would look in grayscale
- Include alternative text when publishing online

### Common Pitfalls

Let's look at some common mistakes to avoid:

#### 1. Poor Scale Choices

A common manipulation is truncating the y-axis to exaggerate differences:

```python
# Create sample data
years = [2018, 2019, 2020, 2021, 2022]
revenue = [98.2, 99.8, 100.5, 102.3, 103.7]

# Create two plots for comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: With truncated y-axis (misleading)
ax1.bar(years, revenue, color='steelblue')
ax1.set_title('Revenue Over Time (Misleading)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Revenue ($ millions)')
ax1.set_ylim(95, 105)  # Truncated y-axis

# Plot 2: With y-axis starting at 0 (more honest)
ax2.bar(years, revenue, color='steelblue')
ax2.set_title('Revenue Over Time (Better Representation)')
ax2.set_xlabel('Year')
ax2.set_ylabel('Revenue ($ millions)')
ax2.set_ylim(0, 110)  # Y-axis starts at 0

plt.tight_layout()
plt.show()
```

#### 2. Inappropriate Chart Types

Not all chart types work for all data:

```python
# Sample data: Market Share percentages
companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
market_share = [35, 25, 20, 12, 8]

# Create comparison of pie chart vs bar chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Pie chart (can be harder to compare values)
ax1.pie(market_share, labels=companies, autopct='%1.1f%%', startangle=90)
ax1.set_title('Market Share (Pie Chart)')

# Bar chart (easier to compare values precisely)
ax2.bar(companies, market_share, color='steelblue')
ax2.set_title('Market Share (Bar Chart)')
ax2.set_ylabel('Percentage (%)')
ax2.set_ylim(0, 40)

plt.tight_layout()
plt.show()
```

#### 3. Overloading with Information

Too much information reduces comprehension:

```python
# Generate complex versus simple visualization
np.random.seed(42)
x = np.arange(50)
data1 = x + np.random.normal(0, 5, 50)
data2 = x + np.random.normal(0, 10, 50)
data3 = x + np.random.normal(5, 2, 50)
data4 = x + np.random.normal(-5, 3, 50)
data5 = x + np.random.normal(10, 8, 50)

# Create comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Overloaded chart
ax1.plot(x, data1, label='Series 1')
ax1.plot(x, data2, label='Series 2')
ax1.plot(x, data3, label='Series 3')
ax1.plot(x, data4, label='Series 4')
ax1.plot(x, data5, label='Series 5')
ax1.set_title('Overloaded Visualization')
ax1.legend()
ax1.grid(True)

# Simplified chart focusing on key series
ax2.plot(x, data1, label='Main Series', linewidth=2, color='steelblue')
ax2.plot(x, data3, label='Comparison Series', linewidth=2, color='darkred')
ax2.set_title('Focused Visualization')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

#### 4. Poor Color Choices

Colors should enhance understanding, not confuse:

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [23, 42, 15, 37, 18]

# Create comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Poor color choice (rainbow colors make it hard to interpret)
rainbow_colors = ['red', 'orange', 'yellow', 'green', 'blue']
ax1.bar(categories, values, color=rainbow_colors)
ax1.set_title('Poor Color Choice (Rainbow)')
ax1.set_ylabel('Value')

# Better color choice (sequential or single hue for quantitative data)
better_colors = sns.color_palette("Blues", len(categories))
ax2.bar(categories, values, color=better_colors)
ax2.set_title('Better Color Choice (Sequential)')
ax2.set_ylabel('Value')

plt.tight_layout()
plt.show()
```

## Choosing the Right Visualization

Selecting the appropriate visualization is crucial for effectively communicating your message.

### Data Types and Visualization Methods

Different data types call for different visualization approaches:

#### Categorical Data

- **Bar charts**: Compare values across categories
- **Pie charts**: Show composition (parts of a whole)
- **Count plots**: Display frequency of categories
- **Dot plots**: Alternative to bar charts, good for showing ranking

#### Numerical Data

- **Histograms**: Show distribution of a single variable
- **Box plots**: Display summary statistics and outliers
- **Density plots**: Smooth representation of distribution
- **Violin plots**: Combine box plot and density plot

#### Time Series Data

- **Line charts**: Show trends over time
- **Area charts**: Emphasize cumulative totals over time
- **Calendar heatmaps**: Show patterns by day/week/month
- **Candlestick charts**: Show open/high/low/close values (finance)

#### Relational Data

- **Scatter plots**: Show relationship between two variables
- **Bubble charts**: Show relationship between three variables
- **Heatmaps**: Show relationships between many variables
- **Network graphs**: Show connections between entities

#### Geospatial Data

- **Choropleth maps**: Color regions by value
- **Dot maps**: Place dots on a map
- **Bubble maps**: Bubbles on a map sized by value
- **Flow maps**: Show movement between locations

### Visualization Selection Framework

To choose the right visualization, ask these questions:

1. **What is the purpose of the visualization?**
   - Explore data
   - Explain findings
   - Monitor trends
   - Make comparisons

2. **What type of data do you have?**
   - Categorical
   - Numerical
   - Time series
   - Geospatial
   - Hierarchical
   - Network

3. **What relationship are you trying to show?**
   - Comparison
   - Distribution
   - Composition
   - Correlation
   - Ranking
   - Trend over time
   - Geospatial pattern

Let's create a decision tree to help select the right visualization:

#### 2. Distribution

```python
# Generate sample data for distribution
np.random.seed(42)
data1 = np.random.normal(100, 15, 1000)
data2 = np.random.normal(90, 20, 1000)
data3 = np.random.normal(110, 10, 1000)

# Create histograms for distribution comparison
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(data1, bins=20, alpha=0.5, label='Product A')
ax.hist(data2, bins=20, alpha=0.5, label='Product B')
ax.hist(data3, bins=20, alpha=0.5, label='Product C')

ax.set_title('Distribution of Customer Ratings')
ax.set_xlabel('Rating Score')
ax.set_ylabel('Frequency')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

#### 3. Composition

```python
# Sample data for composition
categories = ['Marketing', 'R&D', 'Operations', 'Sales', 'Administration']
budget_allocation = [25, 30, 20, 15, 10]

# Create pie chart for composition
fig, ax = plt.subplots(figsize=(10, 6))

ax.pie(budget_allocation, labels=categories, autopct='%1.1f%%', startangle=90)
ax.set_title('Budget Allocation by Department')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.tight_layout()
plt.show()
```

```python
# Create stacked bar chart for comparative composition
departments = ['Marketing', 'R&D', 'Operations', 'Sales', 'Admin']
company_a = [20, 35, 15, 20, 10]
company_b = [25, 25, 20, 20, 10]
company_c = [15, 30, 25, 15, 15]

fig, ax = plt.subplots(figsize=(10, 6))

bottom_a = 0
bottom_b = 0
bottom_c = 0

x = ['Company A', 'Company B', 'Company C']
width = 0.6

for i, dept in enumerate(departments):
    ax.bar(x[0], company_a[i], width, bottom=bottom_a, label=dept if i == 0 else "")
    bottom_a += company_a[i]
    
    ax.bar(x[1], company_b[i], width, bottom=bottom_b, label=dept if i == 0 else "")
    bottom_b += company_b[i]
    
    ax.bar(x[2], company_c[i], width, bottom=bottom_c, label=dept if i == 0 else "")
    bottom_c += company_c[i]

# Add legend with custom handles
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=plt.cm.tab10(i), label=dept) for i, dept in enumerate(departments)]
ax.legend(handles=legend_elements, title='Department')

ax.set_title('Budget Allocation Comparison')
ax.set_xlabel('Company')
ax.set_ylabel('Percentage (%)')
ax.set_ylim(0, 100)

for i, company in enumerate([company_a, company_b, company_c]):
    ax.text(i, 105, f'Total: 100%', ha='center')

plt.tight_layout()
plt.show()
```

#### 4. Relationship

```python
# Generate sample data for relationship
np.random.seed(42)
advertising = np.random.uniform(10, 100, 50)
sales = 5 + 2 * advertising + np.random.normal(0, 10, 50)
roi = (sales - advertising) / advertising * 100

# Create scatter plot for relationship
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(advertising, sales, c=roi, s=roi+20, cmap='viridis', alpha=0.7)

ax.set_title('Relationship Between Advertising Spend and Sales')
ax.set_xlabel('Advertising Spend ($ thousands)')
ax.set_ylabel('Sales ($ thousands)')
ax.grid(alpha=0.3)

# Add a color bar to show ROI values
cbar = plt.colorbar(scatter)
cbar.set_label('ROI (%)')

plt.tight_layout()
plt.show()
```

#### 5. Trend Over Time

```python
# Generate sample data for trend over time
np.random.seed(42)
months = pd.date_range(start='2022-01-01', end='2022-12-31', freq='ME')
trend = np.linspace(100, 150, len(months))
seasonal = 20 * np.sin(np.linspace(0, 2*np.pi, len(months)))
noise = np.random.normal(0, 5, len(months))
sales = trend + seasonal + noise

# Create line chart for trend
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(months, sales, marker='o', linestyle='-', linewidth=2, markersize=8)

ax.set_title('Monthly Sales Trend (2022)')
ax.set_xlabel('Month')
ax.set_ylabel('Sales ($ thousands)')
ax.grid(True, alpha=0.3)

# Format x-axis to show month names
ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b'))

plt.tight_layout()
plt.show()
```

## Introduction to Matplotlib

Matplotlib is the foundational visualization library in Python. Understanding its architecture is key to creating effective visualizations.

### Matplotlib Architecture

Matplotlib has a hierarchical structure:

1. **Figure**: The top-level container that holds everything
2. **Axes**: The actual plotting area (a figure can have multiple axes)
3. **Axis**: The number-lines themselves (each axes has an x-axis and y-axis)
4. **Artists**: Everything visible on the figure (lines, text, legends, etc.)

### Figure and Axes

The Figure and Axes objects are central to Matplotlib:

```python
# Creating a basic figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data on the axes
ax.plot(x, y)

# Add labels and title using axes methods
ax.set_title('A Basic Sine Wave')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a grid
ax.grid(True, alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()
```

Multiple subplots can be created within a single figure:

```python
# Creating multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x/5) * np.sin(x)

# Plot on different axes
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Wave')
axs[0, 0].grid(True, alpha=0.3)

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].grid(True, alpha=0.3)

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_ylim(-5, 5)  # Limit y-axis for tangent
axs[1, 0].grid(True, alpha=0.3)

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Damped Sine Wave')
axs[1, 1].grid(True, alpha=0.3)

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```

### Object-Oriented vs. Pyplot Interface

Matplotlib offers two interfaces:

#### 1. Pyplot Interface

This is more concise but less flexible:

```python
# Pyplot interface example
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

#### 2. Object-Oriented Interface

This provides more control and is recommended for complex visualizations:

```python
# Object-oriented interface example
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y1, label='Sine')
ax.plot(x, y2, label='Cosine')
ax.set_title('Sine and Cosine Waves')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

The object-oriented approach is generally preferred because:
- It's more explicit about which figure and axes you're working with

## Basic Plot Types

Let's explore the most common plot types and when to use them.

### Line Plots

Line plots are best for showing trends over time or continuous relationships:

```python
# Basic line plot
fig, ax = plt.subplots(figsize=(10, 6))

# Generate time series data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + np.pi/4)
y3 = np.sin(x + np.pi/2)

# Plot multiple lines
ax.plot(x, y1, label='Series 1', linewidth=2)
ax.plot(x, y2, label='Series 2', linewidth=2)
ax.plot(x, y3, label='Series 3', linewidth=2)

# Add labels and title
ax.set_title('Multiple Line Plot Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add legend and grid
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

Line plot variations:

```python
# Line plot variations
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Generate data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)               # sin function with 100 points

# Basic line plot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Basic Line Plot')
axs[0, 0].grid(True, alpha=0.3)

# Line plot with markers
# The error was here - trying to use different sizes for x and y
# Fix: Using same number of points for both, but we'll use fewer points with markers
axs[0, 1].plot(x[::10], y1[::10], 'o-', markersize=8, linewidth=2)  # Take every 10th point
axs[0, 1].set_title('Line Plot with Markers')
axs[0, 1].grid(True, alpha=0.3)

# Stepped line plot
axs[1, 0].step(x, y1, where='mid')
axs[1, 0].set_title('Stepped Line Plot')
axs[1, 0].grid(True, alpha=0.3)

# Area plot (filled line)
axs[1, 1].fill_between(x, y1, alpha=0.5)
axs[1, 1].plot(x, y1, color='black', alpha=0.8)
axs[1, 1].set_title('Area Plot (Filled Line)')
axs[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

When to use line plots:
- Showing trends over time
- Visualizing continuous data
- Comparing multiple series over the same domain
- Displaying cumulative values

Best practices for line plots:
- Use solid lines for actual data, dashed lines for projections
- Keep the number of lines to a minimum (ideally 3-5 max)
- Use markers sparingly, typically only when data points are significant
- Consider using area plots to emphasize the magnitude

### Bar Charts

Bar charts are ideal for comparing quantities across categories:

```python
# Basic bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [25, 40, 30, 55, 15]

# Create bar chart
ax.bar(categories, values, color='steelblue', edgecolor='black', alpha=0.7)

# Add labels and title
ax.set_title('Basic Bar Chart')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Add grid lines
ax.grid(axis='y', alpha=0.3)

# Add value labels above bars
for i, v in enumerate(values):
    ax.text(i, v + 1, str(v), ha='center')

plt.tight_layout()
plt.show()
```

Bar chart variations:

```python
# Bar chart variations
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Horizontal bar chart
axs[0, 0].barh(categories, values, color='steelblue', edgecolor='black', alpha=0.7)
axs[0, 0].set_title('Horizontal Bar Chart')
axs[0, 0].grid(axis='x', alpha=0.3)

# Grouped bar chart
x = np.arange(len(categories))
width = 0.35
group1 = [20, 35, 30, 35, 27]
group2 = [25, 32, 34, 20, 25]

axs[0, 1].bar(x - width/2, group1, width, label='Group 1')
axs[0, 1].bar(x + width/2, group2, width, label='Group 2')
axs[0, 1].set_title('Grouped Bar Chart')
axs[0, 1].set_xticks(x)
axs[0, 1].set_xticklabels(categories)
axs[0, 1].legend()
axs[0, 1].grid(axis='y', alpha=0.3)

# Stacked bar chart
axs[1, 0].bar(categories, group1, label='Group 1')
axs[1, 0].bar(categories, group2, bottom=group1, label='Group 2')
axs[1, 0].set_title('Stacked Bar Chart')
axs[1, 0].legend()
axs[1, 0].grid(axis='y', alpha=0.3)

# Error bar chart
errors = [3, 5, 2, 4, 3]
axs[1, 1].bar(categories, values, yerr=errors, capsize=10, 
             color='steelblue', edgecolor='black', alpha=0.7)
axs[1, 1].set_title('Bar Chart with Error Bars')
axs[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

When to use bar charts:
- Comparing values across categories
- Showing rankings or counts
- Displaying distributions of discrete data
- Comparing parts to the whole (stacked bars)

Best practices for bar charts:
- Start the y-axis at zero to avoid misleading comparisons
- Order bars logically (e.g., by value, alphabetically, or chronologically)
- Use horizontal bars when category labels are long
- Limit the number of categories to maintain readability

### Histograms and Count Plots

Histograms and count plots show distributions of data:

```python
# Basic histogram
fig, ax = plt.subplots(figsize=(10, 6))

# Generate sample data
data = np.random.normal(100, 15, 1000)

# Create histogram
ax.hist(data, bins=20, color='steelblue', edgecolor='black', alpha=0.7)

# Add labels and title
ax.set_title('Basic Histogram')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')

# Add grid lines
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Histogram and count plot variations:

```python
# Histogram and count plot variations
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Multiple histograms
data1 = np.random.normal(100, 15, 1000)
data2 = np.random.normal(90, 20, 1000)

axs[0, 0].hist([data1, data2], bins=20, alpha=0.7, label=['Dataset 1', 'Dataset 2'])
axs[0, 0].set_title('Multiple Histograms')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].legend()
axs[0, 0].grid(axis='y', alpha=0.3)

# Density histogram (normalized)
axs[0, 1].hist(data, bins=20, density=True, color='steelblue', 
              edgecolor='black', alpha=0.7)
axs[0, 1].set_title('Density Histogram')
axs[0, 1].set_xlabel('Value')
axs[0, 1].set_ylabel('Density')
axs[0, 1].grid(axis='y', alpha=0.3)
# Count plot (bar chart of categorical counts)
categories = ['A', 'B', 'C', 'D', 'E']
counts = [25, 40, 30, 55, 15]

axs[1, 0].bar(categories, counts, color='steelblue', edgecolor='black', alpha=0.7)
axs[1, 0].set_title('Count Plot')
axs[1, 0].set_xlabel('Category')
axs[1, 0].set_ylabel('Count')
axs[1, 0].grid(axis='y', alpha=0.3)

# Cumulative histogram
axs[1, 1].hist(data, bins=20, cumulative=True, color='steelblue', 
              edgecolor='black', alpha=0.7)
axs[1, 1].set_title('Cumulative Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Cumulative Frequency')
axs[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

When to use histograms and count plots:
- Showing the distribution of a continuous variable (histogram)
- Displaying the frequency of categorical data (count plot)
- Comparing distributions across groups
- Identifying outliers and understanding data spread

Best practices for histograms:
- Choose an appropriate number of bins (too few hides detail, too many shows noise)
- Consider density plots for comparing distributions of different sizes
- Include summary statistics (mean, median) when relevant
- Use cumulative histograms to show percentiles

### Scatter Plots

Scatter plots show relationships between two numerical variables:

```python
# Basic scatter plot
fig, ax = plt.subplots(figsize=(10, 6))

# Generate sample data
np.random.seed(42)
x = np.random.normal(50, 10, 100)
y = 0.5 * x + np.random.normal(0, 10, 100)

# Create scatter plot
ax.scatter(x, y, alpha=0.7)

# Add labels and title
ax.set_title('Basic Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add grid lines
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

Scatter plot variations:

```python
# Scatter plot variations
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Basic scatter plot
axs[0, 0].scatter(x, y, alpha=0.7)
axs[0, 0].set_title('Basic Scatter Plot')
axs[0, 0].grid(True, alpha=0.3)

# Scatter plot with varying size
sizes = np.random.randint(10, 100, 100)
axs[0, 1].scatter(x, y, s=sizes, alpha=0.7)
axs[0, 1].set_title('Scatter Plot with Varying Size')
axs[0, 1].grid(True, alpha=0.3)

# Scatter plot with varying color
colors = np.random.rand(100)
scatter = axs[1, 0].scatter(x, y, c=colors, cmap='viridis', alpha=0.7)
axs[1, 0].set_title('Scatter Plot with Varying Color')
axs[1, 0].grid(True, alpha=0.3)
plt.colorbar(scatter, ax=axs[1, 0])

# Scatter plot with regression line
axs[1, 1].scatter(x, y, alpha=0.7)

# Calculate and plot regression line
m, b = np.polyfit(x, y, 1)
x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + b
axs[1, 1].plot(x_line, y_line, color='red', linestyle='--')
axs[1, 1].set_title(f'Scatter with Regression Line (y = {m:.2f}x + {b:.2f})')
axs[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

When to use scatter plots:
- Examining relationships between two numerical variables
- Identifying correlations, clusters, or outliers
- Showing distribution patterns
- Visualizing multiple dimensions using size, color, or shape

Best practices for scatter plots:
- Add a trend line to highlight relationships
- Use transparency (alpha) when plotting many points
- Consider using size or color to show additional dimensions
- Add descriptive annotations for important points

### Colors and Styles

Matplotlib offers many color options and style sheets:

```python
# Exploring color options
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Named colors
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 15]

axs[0, 0].bar(categories, values, color=['red', 'green', 'blue', 'orange', 'purple'])
axs[0, 0].set_title('Using Named Colors')
axs[0, 0].grid(axis='y', alpha=0.3)

# Color map (sequential)
x = np.linspace(0, 10, 5)
for i in range(5):
    axs[0, 1].plot(x, i*x, color=plt.cm.Blues(i/5), linewidth=2, label=f'Line {i+1}')
axs[0, 1].set_title('Sequential Color Map (Blues)')
axs[0, 1].legend()
axs[0, 1].grid(True, alpha=0.3)

# Color map (diverging)
x = np.linspace(0, 10, 100)
y = np.sin(x)
points = axs[1, 0].scatter(x, y, c=y, cmap='RdBu', s=50)
axs[1, 0].set_title('Diverging Color Map (RdBu)')
axs[1, 0].grid(True, alpha=0.3)
plt.colorbar(points, ax=axs[1, 0])

# Comparing style sheets
with plt.style.context('ggplot'):
    axs[1, 1].plot(x, np.sin(x), label='sin(x)')
    axs[1, 1].plot(x, np.cos(x), label='cos(x)')
    axs[1, 1].set_title('Using ggplot Style')
    axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

Available style sheets:

```python
# Create a plot with different style sheets
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
axs = axs.flatten()

styles = ['default', 'seaborn-v0_8-whitegrid', 'ggplot', 'fivethirtyeight', 'bmh', 
          'seaborn-v0_8-darkgrid', 'classic', 'dark_background', 'seaborn-v0_8-poster']

# Generate sample data
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Plot with different styles
for i, style in enumerate(styles):
    with plt.style.context(style):
        axs[i].plot(x, y)
        axs[i].set_title(style)
        axs[i].tick_params(axis='both', which='major', labelsize=8)

plt.tight_layout()
plt.show()
```

### Annotations and Text

Adding annotations and text elements enhances your visualizations:

```python
# Text and annotations example
fig, ax = plt.subplots(figsize=(10, 6))

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic plot
ax.plot(x, y, linewidth=2)

# Add title with custom font size and weight
ax.set_title('Sine Wave with Annotations', fontsize=16, fontweight='bold')

# Add axis labels with custom font properties
ax.set_xlabel('X-axis', fontsize=12, fontstyle='italic')
ax.set_ylabel('Y-axis', fontsize=12, fontstyle='italic')

# Add text at specific coordinates
ax.text(2, 0.7, 'Local Maximum', fontsize=12, ha='center', 
        bbox=dict(facecolor='white', alpha=0.7, boxstyle='round,pad=0.5'))

# Add arrow annotation
ax.annotate('Local Minimum', xy=(5, -1), xytext=(6, -0.5),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
            fontsize=12, ha='center',
            bbox=dict(facecolor='white', alpha=0.7, boxstyle='round,pad=0.5'))

# Add mathematical expressions using LaTeX
ax.text(8, 0.9, r'$f(x) = \sin(x)$', fontsize=14)

# Add grid
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Layouts and Subplots

Complex layouts help tell a more complete story:

```python
# Advanced subplot layouts
# Using GridSpec for more control
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(2, 3, height_ratios=[1, 1.5])

# Main time series plot (spans the top row)
ax1 = plt.subplot(gs[0, :])
x = np.arange(0, 100)
y = np.cumsum(np.random.normal(0, 10, 100))
ax1.plot(x, y, linewidth=2)
ax1.set_title('Cumulative Sum of Random Walk')
ax1.grid(True, alpha=0.3)

# Histogram in bottom-left
ax2 = plt.subplot(gs[1, 0])
ax2.hist(y, bins=15, orientation='horizontal', color='steelblue', alpha=0.7)
ax2.set_title('Distribution')
ax2.grid(True, alpha=0.3)

# Scatter plot in bottom-middle
ax3 = plt.subplot(gs[1, 1])
ax3.scatter(x[:-1], np.diff(y), alpha=0.7)
ax3.set_title('Step Size vs. Position')
ax3.grid(True, alpha=0.3)

# Box plot in bottom-right
ax4 = plt.subplot(gs[1, 2])
ax4.boxplot(y)
ax4.set_title('Summary Statistics')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Multi-dimensional Data

Visualizing data with multiple dimensions:

```python
# Multi-dimensional visualization example
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Generate sample data
np.random.seed(42)
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
sizes = 1000 * np.random.rand(n)
colors = np.random.rand(n)
categories = np.random.choice(['A', 'B', 'C', 'D'], n)

# Bubble chart (3D: x, y, size)
scatter1 = axs[0, 0].scatter(x, y, s=sizes, alpha=0.7)
axs[0, 0].set_title('Bubble Chart')
axs[0, 0].grid(True, alpha=0.3)

# Bubble chart with color dimension (4D: x, y, size, color)
scatter2 = axs[0, 1].scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
axs[0, 1].set_title('Bubble Chart with Color Dimension')
axs[0, 1].grid(True, alpha=0.3)
plt.colorbar(scatter2, ax=axs[0, 1])

# Grouped scatter plot by category (using different colors)
for category, color in zip(['A', 'B', 'C', 'D'], ['red', 'green', 'blue', 'purple']):
    mask = categories == category
    axs[1, 0].scatter(x[mask], y[mask], label=category, color=color, alpha=0.7)
axs[1, 0].set_title('Grouped Scatter by Category')
axs[1, 0].grid(True, alpha=0.3)
axs[1, 0].legend()

# Parallel coordinates plot (for multidimensional data)
from sklearn.preprocessing import MinMaxScaler

# Create a sample dataset with multiple dimensions
multi_data = np.random.rand(10, 4)  # 10 samples, 4 dimensions
scaler = MinMaxScaler()
multi_data_scaled = scaler.fit_transform(multi_data)

# Plot parallel coordinates
dimensions = ['Feature A', 'Feature B', 'Feature C', 'Feature D']
axs[1, 1].set_title('Parallel Coordinates Plot')

# Set the x-axis range
x = list(range(len(dimensions)))
axs[1, 1].set_xticks(x)
axs[1, 1].set_xticklabels(dimensions)

# Plot each sample
for i in range(multi_data_scaled.shape[0]):
    axs[1, 1].plot(x, multi_data_scaled[i], color=plt.cm.tab10(i % 10), alpha=0.7)

axs[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
